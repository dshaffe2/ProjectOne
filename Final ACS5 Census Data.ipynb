```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from census import Census



# Census API Keys
from config import (api_key)


#c = Census(api_key, year=2016) #2011 -2017?
```

```python
years = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016]

for i in years:
    cen_data = Census(api_key, year=i)
    census_data = cen_data.acs5.get(("NAME", "B01003_001E", "B08301_001E", "B08301_003E","B08301_004E", "B08301_010E", "B08301_019E", "B08101_041E", "B08015_001E", "B08015_002E", "B08015_003E"), {'for': 'metropolitan statistical area/micropolitan statistical area:*'})
    census_pd = pd.DataFrame(census_data)
    


# Column Reordering
    census_pd = census_pd.rename(columns={"B01003_001E": "Population",
                                      "B08301_001E": "Transit Total",
                                      "B08301_003E": "Transit Solo Auto",
                                      "B08301_004E": "Transit Carpool",
                                      "B08301_010E": "Transit Public Transp.",
                                      "B08301_019E": "Transit Walked",
                                      "B08101_041E": "Transit Other",
                                      "NAME": "Name", 
                                      "B08015_001E": "Vehicles",
                                      "B08015_002E": "Vehicles (M)",
                                      "B08015_003E": "Vehicles (F)",
                                      "metropolitan statistical area/micropolitan statistical area": "Metro Code"})
    
    
    # Converting object to floats and then calculating % for transit times
    for j in range(0,10):
        census_pd.iloc[:,j] = pd.to_numeric(census_pd.iloc[:,j] )
    
    census_pd['Year'] = i
    census_pd['New Transit Total'] = (census_pd['Transit Solo Auto']) + (census_pd['Transit Carpool']) + (census_pd['Transit Public Transp.']) + (census_pd['Transit Walked']) + (census_pd['Transit Other'])
    #new transit total calculates a new value for the total transit since there seems to be some discrepancies with the total that is given
    
    census_pd['Solo Auto %'] = round(100*census_pd["Transit Solo Auto"]/census_pd['New Transit Total'],1)
    census_pd['Carpool %'] = round(100*census_pd["Transit Carpool"]/census_pd['New Transit Total'],1)
    census_pd['Public Transp. %'] = round(100*census_pd["Transit Public Transp."]/census_pd['New Transit Total'],1)
    census_pd['Walked %'] = round(100*census_pd["Transit Walked"]/census_pd['New Transit Total'],1)
    census_pd['Other %'] = round(100*census_pd["Transit Other"]/census_pd['New Transit Total'],1)
    
    if (i == years[0]):
        austin_data = census_pd[census_pd['Metro Code'] == '12420']
        nash_data = census_pd[census_pd['Metro Code'] == '34980']
        dallas_data = census_pd[census_pd['Metro Code'] == '19100']
        atl_data = census_pd[census_pd['Metro Code'] == '12060']
    else:
        austin_data = austin_data.append(census_pd[census_pd['Metro Code'] == '12420'])
        nash_data = nash_data.append(census_pd[census_pd['Metro Code'] == '34980'])
        dallas_data = dallas_data.append(census_pd[census_pd['Metro Code'] == '19100'])
        atl_data = atl_data.append(census_pd[census_pd['Metro Code'] == '12060'])
#austin 12420
#atl 12060
#dallas 19100
#nash 34980
    
     

axiscols = ['Name', 'Year', 'Population', 'Vehicles', 'Vehicles (M)', 'Vehicles (F)',
        'Transit Solo Auto','Transit Carpool', 'Transit Public Transp.', 'Transit Walked',
         'Transit Other', 'Transit Total',
      'New Transit Total', 'Solo Auto %', 'Carpool %',
       'Public Transp. %', 'Walked %', 'Other %']
```

```python

atl_data = atl_data[axiscols]
atl_data.iloc[:,2:12] = atl_data.iloc[:,2:12].round().astype(int)
atl_data = atl_data.set_index("Name")
atl_data.to_csv("Atlanta_Census.csv")
atl_data.head()
```
