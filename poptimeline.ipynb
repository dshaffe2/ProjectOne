```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
atl_aq=pd.read_csv("C:/Users/ginaf/Annual_Averages_by_cityAtlex.csv")
atl_pop=pd.read_csv("C:/Users/ginaf/Atlanta_Census.csv")
```

```python
x_axis = np.arange(len(atl_aq["Year"]))
tick_locations = [value+0.4 for value in x_axis] 
```

```python
fig, ax1 = plt.subplots(figsize=(20,6))


#color = 'tab:red'
ax1.set_xlabel("Year")
ax1.set_ylabel("Population (in thousands)")
ax1.plot(x_axis, atl_pop["Population"], color='k')
ax1.tick_params(axis='y', labelcolor='k')
ax1.set_xticks(list(range(0, len(atl_aq["Year"]))))
ax1.set_xticklabels(atl_aq["Year"], rotation="vertical")
ax1.axvline(6, ymin=0, ymax=1, color='g')
ax1.axvline(8, ymin=0, ymax=1, color='g')
ax1.axvline(12, ymin=0, ymax=1, color='g')
ax1.axvline(15, ymin=0, ymax=1, color='g')
ax1.axvline(25, ymin=0, ymax=1, color='r')


ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

#color = 'tab:blue'
ax2.set_ylabel("Air Quality")  # we already handled the x-label with ax1
ax2.plot(x_axis, atl_aq["AQI"], color='b')
ax2.tick_params(axis='y', labelcolor='b')


plt.title("Population vs Air Quality: Atlanta")

plt.savefig("C:/Users/ginaf/Atlantapop.png")
```
